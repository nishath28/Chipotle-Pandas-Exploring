```python
# Import the necessary libraries

import pandas as pd
import numpy as ny
```

```python
# Assign it to a variable called drinks.

drinks= pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv')
print(drinks)
```

```python
# Which continent drinks more beer on average? [mean()]
drinks.groupby('continent')['beer_servings'].mean()
```

```python
#For each continent print the statistics for wine consumption. [describe()]
drinks.groupby('continent')['wine_servings'].describe()
```

```python
#Print the mean alcoohol consumption per continent for every column

drinks.groupby('continent')['beer_servings','spirit_servings','wine_servings','total_litres_of_pure_alcohol'].mean()
```

```python
pd.DataFrame(drinks)
```

```python
#Print the median alcoohol consumption per continent for every column


drinks.groupby('continent')['beer_servings','spirit_servings','wine_servings','total_litres_of_pure_alcohol'].median()
```

```python
#Print the mean, min and max values for spirit consumption.

drinks.groupby('spirit_servings').agg((['mean', 'min', 'max']))


```
